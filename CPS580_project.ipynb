```python
# Initialize SparkSession
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("Read CSV from Google Drive") \
    .getOrCreate()
```

```python

from pyspark.sql import SparkSession
from pyspark.sql.functions import lit

file_path = "/content/drive/MyDrive/itc585/passing_leaders_2022.xlsx"

# Step 6: Read Excel file into a Pandas DataFrame
pandas_df = pandas_df = pd.read_excel(file_path)

# Step 7: Convert Pandas DataFrame to PySpark DataFrame
passing_leaders_2022_df = spark.createDataFrame(pandas_df)

passing_leaders_2022_df = passing_leaders_2022_df.withColumn("date", lit("2022"))

# Step 8: Show the DataFrame
passing_leaders_2022_df.show()
```

```python

file_path = "/content/drive/MyDrive/itc585/passing_leaderss_2023.xlsx"

# Step 6: Read Excel file into a Pandas DataFrame
pandas_df = pandas_df = pd.read_excel(file_path)

# Step 7: Convert Pandas DataFrame to PySpark DataFrame
passing_leaders_2023_df = spark.createDataFrame(pandas_df)
passing_leaders_2023_df = passing_leaders_2023_df.withColumn("date", lit("2023"))


# Step 8: Show the DataFrame
passing_leaders_2023_df.show()
```

```python
concatenated_df = passing_leaders_2023_df.union(passing_leaders_2022_df)

concatenated_df.show(n=100)
```

```python
from pyspark.sql.functions import col, avg, year

# Calculate average stats grouped by player and year
player_stats_by_season = concatenated_df.groupBy("player name", "COMP%","xCOMP%", year(col("date"))).agg(
    avg("TT"),
    avg("CAY"),
    avg("IAY"),
    avg("AYD"),
    avg("LCAD"),
    avg("AGG%"),
    avg("AYTS"),
    avg("COMP%"),
    avg("xCOMP%"),
    (col("COMP%") - col("xCOMP%")).alias("Completion % Above Expectation")
)

player_stats_by_season.show()
```

```python
# Filter for efficient passers with high completion percentage and low time to throw
efficient_passers = concatenated_df.where(
    (col("COMP%") > 0.65) & (col("TT") < 2.7)
)

efficient_passers.show()
```

```python
# Calculate average Time to Throw for each player
avg_tt_per_player = concatenated_df.groupBy("player name").agg({"TT": "avg"})

# Calculate average Aggressiveness for each player
avg_agg_per_player = concatenated_df.groupBy("player name").agg({"AGG%": "avg"})

# Join the two dataframes to analyze player performance
player_performance = avg_tt_per_player.join(avg_agg_per_player, "player name")

player_performance.show()
```

```python
# Train a regression model to predict Completion Percentage
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.regression import LinearRegression

# Prepare feature vector
assembler = VectorAssembler(inputCols=["TT", "AGG%", "AYTS"], outputCol="features")
data = assembler.transform(concatenated_df)

# Split data into training and testing sets
train_data, test_data = data.randomSplit([0.8, 0.2])

# Train Linear Regression model
lr = LinearRegression(featuresCol="features", labelCol="COMP%")
model = lr.fit(train_data)

# Make predictions on test data
predictions = model.transform(test_data)

predictions.show()
```

```python
# Compare performance metrics between two seasons
season1_data = concatenated_df.filter(concatenated_df["date"] == 2023)
season2_data = concatenated_df.filter(concatenated_df["date"] == 2022)

# Calculate average Completion Percentage for each season
avg_completion_season1 = season1_data.agg({"COMP%": "avg"}).collect()[0][0]
avg_completion_season2 = season2_data.agg({"COMP%": "avg"}).collect()[0][0]

# Compare average Completion Percentage between seasons
completion_comparison = avg_completion_season2 - avg_completion_season1
print(completion_comparison)
```
